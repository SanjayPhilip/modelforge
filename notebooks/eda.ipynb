# 📊 California Housing EDA

Exploratory Data Analysis for the California Housing dataset used in ModelForge.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
```

## 1. Load Data

```python
data = fetch_california_housing(as_frame=True)
df = data.frame
df.head()
```

## 2. Data Overview

```python
print(f'Shape: {df.shape}')
print(f'
Columns: {list(df.columns)}')
df.describe()
```

## 3. Target Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MedHouseVal'], kde=True, bins=50)
plt.title('Distribution of Median House Values')
plt.xlabel('Median House Value ($100,000s)')
plt.show()
```

## 4. Correlation Matrix

```python
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

## 5. Geographic Distribution

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['Longitude'], df['Latitude'], \n                     c=df['MedHouseVal'], cmap='viridis', alpha=0.5)
plt.colorbar(scatter, label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('California Housing Prices by Location')
plt.show()
```
